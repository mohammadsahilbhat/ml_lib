# Preprocessing _ by creating pipe

```python
import numpy as np
from ml_lib.preprocessing.scaler import StandardScaler
from ml_lib.preprocessing.imputer import SimpleImputer
from ml_lib.preprocessing.polynomial_features import PolynomialFeatures 
from ml_lib.preprocessing.pipeline import Pipeline
```

- define simple data 

```python
X = np.array([
    [1.0, 2.0],
    [2.0, np.nan],
    [3.0, 6.0],
    [np.nan, 8.0]
])
```

- making pipe

```python
pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('poly_features', PolynomialFeatures(degree=2, include_bias=False))
])
```

```python
X_transformed = pipe.fit_transform(X)
print(X_transformed)
```
