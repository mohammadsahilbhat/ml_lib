# Neural Network Demo

- Imports

```python
import numpy as np

from ml_lib.neural_network.sequential import Sequential

from ml_lib.neural_network.layers import Dense
from ml_lib.neural_network.losses import BinaryCrossEntropy
```

- Data

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])
```

Model

```python
model= Sequential([
     Dense(8,activation= 'relu'),
     Dense(1,activation='Sigmoid')


])

model.compile(loss=BinaryCrossEntropy(),lr=0.1)
model.fit(X,y,epochs=2000,verbose=100)

preds= model.predict(X)
print((preds>0.5).astype(int))
```
