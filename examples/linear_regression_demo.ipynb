# Linear Regression Demo 

- import modules

```python
import numpy as np 
from ml_lib.linear_models import LinearRegression
```

### simple dataset

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])
```

- create model

```python
model = LinearRegression(lr=0.01, epochs=1000)
```

### Train Model

```python
model.fit(X, y)
```

### Predictions

```python
predictions = model.predict(X)
print("Predictions:", predictions)
print("Weights:", model.W)
print("Bias:", model.b)
```

```python
from ml_lib.utils.visuals.regression_plots import regression_dashboard,plot_regression_line

regression_dashboard(X, y, model)
plot_regression_line(X, y, model)
```
