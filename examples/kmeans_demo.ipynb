# Example of Kmeans 

- Importing libraries and model

```python
import numpy as np
import matplotlib.pyplot as plt
from ml_lib.clustering.kmeans import KMeans
```

- defining some data set

```python
X = np.array([
    [1, 2], [1.5, 1.8], [5, 8],
    [8, 8], [1, 0.6], [9, 11]
])
```

- fitting the data

```python
kmeans = KMeans(n_clusters=2)
labels=kmeans.fit(X)
```

- predict again (0r on new data)

```python
pred = kmeans.predict(X)
```

- visualize the clustering

```python
plt.scatter(X[:,0],X[:,1],c=pred ,s=100,cmap='viridis')
plt.scatter(kmeans.centroids[:,0],kmeans.centroids[:,1], c='black', s=200, alpha=0.5,marker='X')
plt.show()
```
