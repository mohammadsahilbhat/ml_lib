# Logistic Regression Demo

- import modules

```python
import numpy as np
from ml_lib.linear_models import LogisticRegression
```

- simple dataset

```python
X = np.array([[1],
              [2],
              [3],
              [6],
              [7],
              [8]])

y = np.array([0, 0, 0, 1, 1, 1])
```

- create model

```python
model = LogisticRegression(lr=0.01, epochs=1000)
```

- Train model

```python
model.fit(X,y)
```

- Predictions

```python
probabilities = model.predict_proba(X)
predictions = model.predict(X)
print("Predicted probabilities:", probabilities)
print("Predicted classes:", predictions)

print("Model weights:", model.W)
print("Model bias:", model.b)
```

## Trying to visualize Sigmoid Curve
   using matplotlib 

```python
import matplotlib.pyplot as plt

x_vals = np.linspace(0, 10, 100).reshape(-1, 1)
y_vals = model.predict_proba(x_vals)

plt.scatter(X, y, label="Data")
plt.plot(x_vals, y_vals, color="red", label="Sigmoid curve")
plt.xlabel("X")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
