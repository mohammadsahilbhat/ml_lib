# Linear Regression Test 

### Imports 


```python
import numpy as np
import pandas as pd
from ml_lib.linear_models.linear_regression import LinearRegression
from ml_lib.preprocessing.imputer import SimpleImputer
from ml_lib.preprocessing.scaler import StandardScaler
```

- read the data

```python
df = pd.read_csv("C:\project\datasets\Linear_Regression_Train.csv")

X = df.drop("target", axis=1).values
y = df["target"].values
```

- checking shapes of data

```python
print(X.shape, y.shape)
print(df.head())
```

- Preprocess The Data

```python
scaler= StandardScaler()

X= scaler.fit_transform(X)

# if missing values exist

imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(X)
```

- Train-Test Split

```python
from  ml_lib.utils.data import train_test_split
X_train, y_train,X_test, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
```

### Evaluate the model

```python
from ml_lib.metrics.math import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R2 Score: {r2}")
```

## Predict for the Given New data 

```python
df1= pd.read_csv(r"C:/project/datasets/Linear_Regression_Test.csv")
X_Test = df1.values
X_Test= scaler.fit_transform(X_Test)
# if missing values exist
X_Test = imputer.fit_transform(X_Test)

y_pred_test = model.predict(X_Test)
print("Predictions on test data:")

df1['Predicted_Target'] = y_pred_test

df1.to_csv("C:\project\datasets\processed\Linear_Regression_Predictions.csv", index=False)

print("predictions saved to Linear_Regression_Predictions.csv")
```
