# Binary Classification Problem


### By using Logistic Regression Model


- Imports

```python
import pandas as pd
import numpy as np
from ml_lib.linear_models.logistic_regression import LogisticRegression
from ml_lib.preprocessing.scaler import StandardScaler
from ml_lib.utils.data import train_test_split
from ml_lib.preprocessing.pipeline import Pipeline
from ml_lib.preprocessing.imputer import SimpleImputer
```

- Load Data

```python
train_df=pd.read_csv(r"C:/project/datasets/train_binary.csv")
test_df= pd.read_csv(r"C:/project/datasets/test_binary.csv")

X= train_df.iloc[:,:-1].values
y= train_df.iloc[:,-1].values

print(f"shape  of X : {X.shape} , shape of y : {y.shape}")


X_test = test_df.values
```

- Preprocessing

```python
pipe = Pipeline([
    ('imputer',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
    ])


X_trans= pipe.fit(X)

X_train,y_train,X_val,y_val= train_test_split(X_trans,y)
```

Creating the model

```python
model = LogisticRegression(
    lr=0.005,
    epochs=3000
)

model.fit(X_train,y_train)
```

- Check Accurracy for the validation set of data

```python
y_val_pred= model.predict(X_val)
val_acc=np.mean(y_val_pred==y_val)*100
print(print("Validation Accuracy:", round(val_acc , 2), "%"))
```

### Finakl prediction for test data

```python
y_test_pred=model.predict(X_test)





test_df["Predicted_Label"]=y_test_pred

test_df.to_csv(r"C:/project/datasets/processed/binary_classification.csv")
```
